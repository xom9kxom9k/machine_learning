1. Генерация данных

```python
import numpy as np
import pandas as pd

x1 = np.linspace(0, 10, 500) 
x2 = np.linspace(0, 5, 500)

y = np.cos(x1) + (x2)**3

df = pd.DataFrame({
    "x1": x1,
    "x2": x2,
    "y": y
})

df.to_csv("data.csv", index=False)
```

2. Открыть файл и построить график

```python
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

x2_const = df["x2"].iloc[100] 
x1_const = df["x1"].iloc[200]

plt.figure()
plt.scatter(df["x1"], np.cos(df["x1"]) + (x2_const)**3, s=10, c="blue")
plt.title(f"y(x1), при x2 = {x2_const:.2f}")
plt.xlabel("x1")
plt.ylabel("y")
plt.grid(True)
plt.show()

plt.figure()
plt.scatter(df["x2"], np.cos(x1_const) + (df["x2"])**3, s=10, c="red")
plt.title(f"y(x2), при x1 = {x1_const:.2f}")
plt.xlabel("x2")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

3. Вывод для каждого столбца

```python
print("Статистика по данным:")
print(df.describe()[["x1", "x2", "y"]].loc[["mean", "min", "max"]])
```

4. Сохранение с новыми условиями

```python
x1_mean = df["x1"].mean()
x2_mean = df["x2"].mean()

filtered = df[(df["x1"] < x1_mean) | (df["x2"] < x2_mean)]
filtered.to_csv("filtered_data.csv", index=False)
```

5. 3D график функции

```python
from mpl_toolkits.mplot3d import Axes3D

X1 = np.linspace(0, 10, 50)
X2 = np.linspace(0, 5, 50)
X1, X2 = np.meshgrid(X1, X2)

Y = np.cos(X1) + (X2)**3

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X1, X2, Y, cmap="viridis")

ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
plt.title("3D график y(x1, x2)")
plt.show()
```
